```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
#Cau 4

year = [1972, 1982, 1992, 2002, 2012]
VietNam = [120, 178, 205, 395, 724]
USA = [10, 25, 58, 119, 274]
plt.plot(year, VietNam, color ='r', label ='Viet Nam', marker='*', linestyle='dashed', linewidth=3, markersize=8)
plt.plot(year, USA, color ='blue', label ='Hoa Kỳ', marker='d', linestyle='dotted', linewidth=3, markersize=8)
plt.xlabel('Năm')
plt.ylabel('Lượng điện tiêu thụ')
plt.title('So sánh lượng tiêu thụ điện')
plt.legend()
plt.show()
```

```python
# Cau 5

MonHoc = ['CSDL', 'QTM', 'MMT','PTDL', 'LTCSDL']
Soluong = [20, 55, 30, 43, 70]
plt.bar(MonHoc, Soluong, color ='b', width = 0.4) #with là độ rộng của bar
plt.xlabel("Môn học")
plt.ylabel("Số lượng SV đăng ký")
plt.title("SV đăng ký HK II")
plt.show()
```

```python
# Cau 5 - ngang

MonHoc = ['CSDL', 'QTM', 'MMT','PTDL']
Soluong = [20, 55, 30, 43]
plt.barh(MonHoc, Soluong, color ='r', height = 0.4) #with là độ rộng của bar
plt.xlabel("Môn học")
plt.ylabel("Số lượng SV đăng ký")
plt.title("SV đăng ký HK II")
plt.show()
```

```python
# Cau 5 - so sanh

X = ['Nhóm A','Nhóm B','Nhóm C','Nhóm D', 'Nhóm E']
nu = [10,20,20,40, 14]
nam = [20,30,25,30, 16]
X_axis = np.arange(len(X)) # Lấy các giá trị để hiển thị các nhóm
plt.bar(X_axis - 0.2, nu, width=0.4, label = 'Nữ', color='orange')
plt.bar(X_axis + 0.2, nam, width=0.4, label = 'Nam', color='g')
plt.xticks(X_axis, X) # X_axis là vị trí, X là tên sẽ hiển thị
plt.xlabel("Nhóm")
plt.ylabel("Số lượng sinh viên")
plt.title("Số sinh viên của từng nhóm")
plt.legend()
plt.show()
```

```python
# Cau 6

data = np.random.randn(100000)
plt.hist(data, bins=50, color='pink', edgecolor='black')
plt.xlabel('Số lượng')
plt.ylabel('Tần suất')
plt.title('Biểu đồ Histogram')
plt.show()
```

```python
# Cau 6 - them seaborn

data = np.random.randn(100000)
sns.histplot(data, bins=50, kde=True, color='b', edgecolor='black')
plt.xlabel('Số lượng')
plt.ylabel('Tần suất')
plt.title('Biểu đồ Histogram')
plt.show()
```

```python
# Cau 7 - Scatter

x = np.array([5,7,8,7,2,17,2,9])
y = np.array([99,86,87,88,111,86,103,87])
plt.scatter(x, y, s = 50, color='r')
x = np.array([2,2,8,1,15,8,12,9,7,11])
y = np.array([100,105,84,105,90,99,90,91,80,85])
plt.scatter(x, y, s = 100, color='g')
plt.show()
```

```python
# Cau 7 - colormap

x = np.array([5,7,8,7,2,17,2,9])
y = np.array([99,86,87,88,111,86,103,87])
colors = np.array([20, 30, 40, 45, 50, 55, 60, 70])
plt.scatter(x, y, c=colors, cmap='Spectral', s=100)
plt.colorbar()
plt.show()
```

```python
# Cau 7  - tham so nang cao

x = np.array([9,5,7,8,7,2,17,2,9,12,32,45,21,45,10, 8, 0, 45, 21, 33])
y = np.array([99,86,87,88,111,86,103,87,33,45,76,87,65,80,77, 5, 67, 89, 23, 9])
colors = np.random.rand(20)
sizes = 100 * np.random.rand(20)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='gnuplot')
plt.title("Tiêu đề biểu đồ Scatterplot")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.colorbar(label='Bảng màu')
plt.show()
```

```python
# Cau 8

monhoc = ['CSDL', 'MMT', 'QTM','AI', 'KTLT', 'CSLT', 'LTCSDL']
data = [67, 105, 35, 52, 43, 41, 100]
plt.pie(data, labels = monhoc)
plt.show()
```

```python
# Cau 8 - nang cao

monhoc = ['CSDL', 'MMT', 'QTM','AI', 'KTLT', 'CSLT', 'LTCSDL']
data = [67, 105, 35, 52, 43, 41, 100]
mausac = ["black", "hotpink", "b", "#4CAF50","yellow","red", "pink"]
vitri = [0.3, 0, 0, 0, 0, 0, 0]
plt.pie(data, labels = monhoc, colors = mausac, explode = vitri, shadow = True)
plt.legend(bbox_to_anchor=(1,0), loc="upper right")
plt.show()
```

```python
# Cau 9

np.random.seed(100)
data = np.random.normal(100, 20, 200)
plt.boxplot(data) # Tham số vert=False sẽ vẽ boxplot nằm ngang
plt.show()
```

```python
# Cau 9 - nhieu box plot

data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
fig = plt.figure(figsize =(7, 5)) #Tạo 1 figure với kích thươc 5x4
ax = fig.add_axes([0, 0, 1, 1])
ax.boxplot(data, vert=False)
plt.show()
```

```python
# Cau 10 - nhieu bieu do trong mot figure

x = np.arange(1,9)
y = x*3
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[2,4,9,16])
plt.title("Hình 1")
plt.subplot(2,1,2)
plt.plot(x,y,"r^")
plt.title("Hình 2")
plt.show()
```

```python
# Cau 10

fig, bieudo = plt.subplots(2, 2)
bieudo[0, 0].plot([1, 2, 3], [4, 5, 6]) #Vẽ tại vị trí dòng 0, cột 0
bieudo[0, 1].scatter([1, 2, 3], [4, 5, 6]) #Vẽ tại vị trí dòng 0, cột 1
bieudo[1, 0].bar([1, 2, 3], [4, 5, 6]) #Vẽ tại vị trí dòng 1, cột 0
bieudo[1, 1].hist([1, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5]) #Vẽ tại vị trí dòng 1, cột 1
plt.show()
```
